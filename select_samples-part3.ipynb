## Conclussion of this
Seems like drug information is not enough to explain the underexpression of EGFR...  

Paragraph 4 of this short paper: https://onlinelibrary.wiley.com/doi/pdf/10.1002/ijc.26434 -- "Profiling of EGFR mRNA and protein expression in 471 breast cancers compared with 10 normal tissues: a candidate biomarker to predict EGFR inhibitor effectiveness" -- says (emphasis is mine):  
"**Overall, EGFR mRNA was markedly underexpressed in the breast tumors relative to the normal breast tissues** (HR+/ ERBB2+: 80%; HR+/ERBB2-: 89%; HR-/ERBB2+: 73%;HR-/ERBB2-: 42%; Tables 1 and 2). EGFR mRNA was overexpressed in only 11% of TNBC (HR-/ERBB2-) carcinomas and in almost none of the tumors in the other three
subgroups (Tables 1 and 2)."  
So perhaps this is okay?
