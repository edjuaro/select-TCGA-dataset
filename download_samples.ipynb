## Selecting only normal samples (there are more tumor than matched normal)

```python
import numpy as np
# patients = [sample.strip('-11') if sample.endswith('-11') else '' for sample in full_dataset]
# patients = [sample.replace('-11','') if sample.endswith('-11') else '' for sample in full_dataset]
patients = [sample[:12] if sample.endswith('-11') else '' for sample in full_dataset]
patients = np.unique(sorted([x for x in patients if x !='']))
print("There are {} patients with normal samples".format(len(patients)))
print(patients)
```

## Selecting samples with the same IDs as the ones from normal
Effectively selecting all tummor-matched_normal pairs. Except there are some anomalies

```python
new_patients = [a if a[:-3] in patients else '' for a in list(full_dataset)]
new_patients = np.unique(sorted([x for x in new_patients if x !='']))

print("There are {} patients now, there should be {} (twice as much as the cell above. Potentially removing Normal Samples without the corresponding tumor.)".format(len(new_patients), 2*len(patients)))
# Some samples have only normal tissue (for whatever reason)
to_remove = []
i = 0
j = 0
for sample in new_patients:
    if sample.endswith('-11'):
        i+=1
    elif sample.endswith('-01'):
        j+=1
    if sample.replace('-11','-01') not in new_patients:
        to_remove.append(sample)
        print('\tSample "{}" will be removed'.format(sample))
        i-=1

new_patients = [x for x in new_patients if x not in to_remove]
print("There are {} patients now ({} are normal and {} are tumor).".format(len(new_patients),i,j))
```

## Selecting only samples with tummor-matched_normal pairs

```python
all_matched_samples = full_dataset[new_patients]
all_matched_samples
```

```python
import cuzcatlan as cusca
cusca.df2gct(all_matched_samples, use_index=True, name='all_matched_samples.gct', extra_columns=0)
```
